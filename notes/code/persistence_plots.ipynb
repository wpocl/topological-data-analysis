```python
import numpy as np
import gudhi as gd
import matplotlib.pyplot as plt

from helpers import sample_from_torus
```

```python
num_points = 1000
major_radius = 5
minor_radius = 2

x, y, z = sample_from_torus(num_points, major_radius, minor_radius)
torus = np.squeeze(np.dstack((x, y, z)))
```

```python
rips = gd.RipsComplex(points=torus, max_edge_length=3)
st = rips.create_simplex_tree(max_dimension=2)
```

```python
diag = st.persistence(min_persistence=0.8)
gd.plot_persistence_barcode(diag)
plt.savefig("figures/torus_barcode.png", bbox_inches="tight")
```

```python
gd.plot_persistence_diagram(diag)
plt.savefig("figures/torus_persistence_diagram.png", bbox_inches="tight")
```
