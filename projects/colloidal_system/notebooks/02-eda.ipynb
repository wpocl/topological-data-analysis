```python
import pickle

import pandas as pd
import numpy as np


import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

```python
with open("data/02-preprocessed/data.pickle", "rb") as file:
    data = pickle.load(file)
```

```python
key = pd.read_pickle("data/02-preprocessed/key.pickle")
```

```python
# Create a figure with 12 subplots
fig, axs = plt.subplots(3, 4, figsize=(15, 10))

# Flatten the axs array for easy iteration
axs = axs.flatten()

# Iterate over the subplots and plot scatter plots
for i, ax in enumerate(axs):
    ax.scatter(x=data[2 * i, 0, :, 0], y=data[2 * i, 9, :, 1], s=6)

    n_b = round(key.columns.get_level_values(1)[2 * i], 1)

    ax.set_title(f'n_b: {n_b}')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')

# Add a title to the figure
t = key.index[0]
plt.suptitle(f"t: {t}", fontsize=20)

# Adjust the spacing between subplots
plt.tight_layout()
```

```python
# Create a figure with 12 subplots
fig, axs = plt.subplots(3, 4, figsize=(15, 10))

# Flatten the axs array for easy iteration
axs = axs.flatten()

# Iterate over the subplots and plot scatter plots
for i, ax in enumerate(axs):
    ax.scatter(x=data[2 * i, -1, :, 0], y=data[2 * i, -1, :, 1], s=6)

    n_b = round(key.columns.get_level_values(1)[2 * i], 1)

    ax.set_title(f'n_b: {n_b}')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')

# Add a title to the figure
t = key.index[-1]
plt.suptitle(f"t: {t}", fontsize=20)

# Adjust the spacing between subplots
plt.tight_layout()
```
